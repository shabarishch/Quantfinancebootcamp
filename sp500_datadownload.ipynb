```python
import yfinance as yf
import time
import pandas as pd
import json
import os
```

```python
symbols = pd.read_html(
    "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
)[0]["Symbol"].tolist()
```

```python
symbols
```

```python
for symbol in symbols:
    data = yf.Ticker(symbol)
    #data_df = data.history(period='500y')
    #data_df.to_csv(f'Data/{symbol}.csv')
    with open(f"StockDetails/{symbol}.json", "w") as f:
        json.dump(data.info, f)
    time.sleep(2)
```
